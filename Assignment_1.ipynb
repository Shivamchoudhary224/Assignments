Q1. What are the three measures of central tendency?

Ans. Three Measures of central tendency are:

1. Mean: It is the sum of the values of all data points divided by the total number of data points.

2. Median: It is the center most value of the dataset when it is ordered .

3. Mode: It is the most occuring value in a dataset.

Q5. How are measures of dispersion such as range, variance, and standard deviation used to describe
the spread of a dataset? Provide an example.

Ans. 1. Range: It is the spread of data from lowest to the heighest value of a data.

example: [10,20,30,40,50]

R(range)=H(heighest)-L(lowest)

R = 50-10 = 40

2. Variance: It explains the degree of spread in data, it is calculated by taking the square of average deviation from mean.

variance is of two types i.e population variance and sample variance the only difference between them is we divide sample variance by (n-1), reason being that it is observed that when we divide by n-1 our sample mean tends to move towards population mean.

example: [10,20,30,40,50]

variance = 200

3. Standard deviation: It shows how much away data is from the centre of distribution.
It is calculated by taking square root of variance.

example: [10,20,30,40,50]

Standar deviation: 14.142135623730951

Q8. What do you understand about skewness in data?

Ans: Skewness of data is a property of distribution which refers to a situation where out data is not equally distributed on the both side of mean. When tail of data is more extended towards right side it refers to as right skewed data/ positively skewed data, where as when tail is more extended toward left side of the mean then it is called as left skewed / negatively skewed data.
