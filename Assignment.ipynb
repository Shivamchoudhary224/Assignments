Q3. What are the different types of data and how do they differ from each other? Provide an example of
each type of data.

Ans. Data is of two types:

1. Qualitative Data: It has categorial data. For eg: Gender of people in an office it can be male, female, LGBTQ.

It further has 2 types:

 1. Nominal data: It is a type of data in which rank does not matter,Example: Size of clothes.

 2. Ordinal data: It is a type of data in which rank plays an important role. Example: Rating given to a resturant.

2. Quantative Data: It has only numerical data. Eg: Height of 100 people.

It further has 2 types:

 1. Discrete data: It has data whose values are always whole numbers. Eg: number of children.

 2. Continous Data: It has data whose values can take any numerical value. Eg: weight.

Q5. Explain the concept of levels of measurement and give an example of a variable for each level.

Ans. Levels of measurment:
1. Nominal Scale data: In this data can only be categorize.
Example: gender related data.

2. ordinal Scale data: Here data can be categorize and ranked. Example Rating given at a resturant.

3. Interval scale data: Here data can be categorize, rank and also can be infered into intervals but cannot has 0 as its starting point. Example: Test score

4. Ratio Scale data: Here data can be categorize, rank and also can be infered into intervals but can has 0 as its starting point. Example: height

Q10. What are some common measures of central tendency and variability used in statistics? Explain
how each measure can be used to describe a dataset.

Ans. For central tendancy, we have mean, median and mode. Mean is the sum of all values divided by total number of data points but mean is more vulnerable to outliers thats why we use median which is insenstive to outliers as it is the mid point of the distribution, where as mode is used when we have categorical data, it the most frequently occuring element.

for Variability we have variance and standard deviations, variance explains the spread of data where as standard deviation explains how far each data value is from mean.
