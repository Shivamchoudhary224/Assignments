Q8: What is the z score? State the importance of the z score.

Ans: z-score, also known as the standard score, is a statistical measure that quantifies how far a particular data point is from the mean of a dataset in terms of standard deviations. It allows to standardize and compare data points from different datasets.

Z score allows us to perform hypothesis testing, standardizing the distributions and identifying outliers.
